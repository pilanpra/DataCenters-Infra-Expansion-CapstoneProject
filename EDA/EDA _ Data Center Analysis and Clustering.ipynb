```python
import matplotlib.pyplot as plt
import seaborn as sns

# 📊 Step 1: Ensure certification columns are categorical
cert_columns = ['ISO27001_CERTIFIED', 'HIPAA', 'SOC_2_TYPE_II_CERTIFIED']

for col in cert_columns:
    df[col] = df[col].astype(str)  # Ensure all values are strings
    df[col] = df[col].replace({'Yes': 'Certified', 'No': 'Not Certified'})  # Rename for clarity

# ✅ Check if conversion worked
print(df[cert_columns].head())  

# 📊 Step 2: Create Swarm Plot for Energy vs. Certification
plt.figure(figsize=(12, 6))

for i, cert in enumerate(cert_columns, 1):
    plt.subplot(1, 3, i)
    sns.swarmplot(data=df, x=cert, y='ENERGY', palette=['#D98880', '#2874A6'])
    plt.title(f"Energy Use Distribution by {cert}", fontsize=14, fontweight='bold')
    plt.xlabel(cert)
    plt.ylabel("Energy (MW)")

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Copy your current working df
df_cluster_ready = df.copy()

# Step 2: Clean numeric columns
df_cluster_ready['ENERGY'] = df_cluster_ready['ENERGY'].astype(str).str.replace(r'[^0-9.]', '', regex=True)
df_cluster_ready['ENERGY'] = pd.to_numeric(df_cluster_ready['ENERGY'], errors='coerce')

df_cluster_ready['AREA'] = df_cluster_ready['AREA'].astype(str).str.replace(r'[^0-9]', '', regex=True)
df_cluster_ready['AREA'] = pd.to_numeric(df_cluster_ready['AREA'], errors='coerce')

df_cluster_ready['IT EQUIPMENT POWER'] = pd.to_numeric(df_cluster_ready['IT EQUIPMENT POWER'], errors='coerce')
df_cluster_ready['PUE'] = pd.to_numeric(df_cluster_ready['PUE'], errors='coerce')

# Step 3: Define features and drop NA
features = ['ENERGY', 'AREA', 'IT EQUIPMENT POWER', 'PUE']
df_cluster_ready = df_cluster_ready.dropna(subset=features)

# Step 4: Scale and cluster
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cluster_ready[features])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster_ready['Cluster'] = kmeans.fit_predict(scaled_features)

# Step 5: PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
df_cluster_ready['PCA1'] = pca_result[:, 0]
df_cluster_ready['PCA2'] = pca_result[:, 1]

# Step 6: Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cluster_ready,
    x='PCA1',
    y='PCA2',
    hue='Cluster',
    palette='coolwarm',
    edgecolor='black',
    alpha=0.8
)
plt.title("📊 Data Center Clusters (via PCA)", fontsize=16, fontweight='bold')
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Copy dataframe for clustering
df_cluster_ready = df.copy()

# Convert numeric columns if needed
df_cluster_ready['ENERGY'] = pd.to_numeric(df_cluster_ready['ENERGY'], errors='coerce')
df_cluster_ready['AREA'] = pd.to_numeric(df_cluster_ready['AREA'], errors='coerce')
df_cluster_ready['IT EQUIPMENT POWER'] = pd.to_numeric(df_cluster_ready['IT EQUIPMENT POWER'], errors='coerce')
df_cluster_ready['PUE'] = pd.to_numeric(df_cluster_ready['PUE'], errors='coerce')

# Drop missing values
features = ['ENERGY', 'AREA', 'IT EQUIPMENT POWER', 'PUE']
df_cluster_ready = df_cluster_ready.dropna(subset=features)

# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cluster_ready[features])

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_cluster_ready['Cluster'] = kmeans.fit_predict(scaled_features)

# PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
df_cluster_ready['PCA1'] = pca_result[:, 0]
df_cluster_ready['PCA2'] = pca_result[:, 1]

# Plot clusters
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cluster_ready,
    x='PCA1',
    y='PCA2',
    hue='Cluster',
    palette='coolwarm',
    edgecolor='black',
    alpha=0.8
)
plt.title("📊 Data Center Clusters (via PCA)", fontsize=16, fontweight='bold')
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.tight_layout()
plt.show()
```
