```python
df_cleaned = pd.read_csv('/Users/prasadpilankar/Documents/BAN/DE2025/DataCenterInfra/clean-data/DC-CleanedData.csv')
df_cleaned = df_cleaned.fillna(df_cleaned.mean())
df_cleaned
```

```python
df_cleaned.columns
```

```python
features = ['ENERGY', 'AREA', 'IT EQUIPMENT POWER',
       'PUE', 'FULL_CABINETS', 'PARTIAL_CABINETS', 'SHARED_RACKSPACE', 'CAGES',
       'SUITES', 'BUILD_TO_SUIT', 'FOOTPRINTS', 'REMOTE_HANDS',
       'YEAR_OPERATIONAL', 'INTERNET_EXCHANGE_POINTS'
]

df_cluster = df_cleaned[features]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cluster)
```

```python
# -----------------------------
# 2. Determine Optimal Clusters (Elbow Method)
# -----------------------------
inertia = []
K_range = range(1, 10)  # Try cluster sizes from 1 to 9

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
# -----------------------------
# 3. Perform K-Means Clustering
# -----------------------------
# Choose optimal K from the Elbow Method (e.g., 3 clusters)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_cluster["Cluster"] = kmeans.fit_predict(df_scaled)
```

```python
# -----------------------------
# 4. Perform PCA for Visualization
# -----------------------------
pca = PCA(n_components=2)  # Reduce to 2D for visualization
df_pca = pca.fit_transform(df_scaled)
df_cluster["PCA1"] = df_pca[:, 0]
df_cluster["PCA2"] = df_pca[:, 1]
```

```python
# Scatter plot of clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_cluster["PCA1"], y=df_cluster["PCA2"], hue=df_cluster["Cluster"], palette="viridis", s=100)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("K-Means Clustering of Data Centers")
plt.legend(title="Cluster")
plt.show()
```
